```python
import pandas as pd
dataset=pd.read_excel('data.xlsx')
dataset=dataset.set_index('Tid')
dataset['Income']= dataset['Income'].str.strip('K')
dataset['Income']=dataset['Income'].astype('int')
dataset
```

```python
import math 
def prob_cont(Income,mean,var):
    return((1/math.sqrt(2*math.pi*var))*(math.exp(-(((Income-mean)**2)/(2*var)))))
```

```python
def Bayes(refund,status,income):
    
    
    # for Evade = Yes
    datasetyes = dataset[dataset['Evade'] == 'Yes']
   
    data_refund1 = datasetyes[datasetyes['Refund'] == refund]
    data_status1 = datasetyes[datasetyes['Marital Status'] == status]
    
    yes_var = datasetyes['Income'].var() 
    yes_mean = datasetyes['Income'].mean()

    prob_of_yes = datasetyes.shape[0] / dataset.shape[0] # yes
    prob_of_refund1 = data_refund1.shape[0] / datasetyes.shape[0]  # refund yes/no
    prob_of_status1 = data_status1.shape[0] / datasetyes.shape[0]  # status married/single/divorced
    prob_of_income1 = prob_cont(income,yes_mean,yes_var) # 90/60
    totalprob1=(prob_of_refund1*prob_of_status1*prob_of_income1)*prob_of_yes
    
    #for Evade = No
    datasetno = dataset[dataset['Evade'] == 'No']
    
    data_refund2 = datasetno[datasetno['Refund'] == refund] 
    data_status2 = datasetno[datasetno['Marital Status'] == status]
    
    no_var = datasetno['Income'].var() 
    no_mean = datasetno['Income'].mean()
    
    prob_of_no = datasetno.shape[0] / dataset.shape[0] # yes/no
    prob_of_refund2 = data_refund2.shape[0] / datasetno.shape[0]  # refund yes/no
    prob_of_status2 = data_status2.shape[0] / datasetno.shape[0]  # status married/single/divorced
    prob_of_income2 = prob_cont(income,no_mean,no_var)# 90/60
   
    totalprob2=(prob_of_refund2*prob_of_status2*prob_of_income2)*prob_of_no

    if(totalprob2>totalprob1):
        print("The Evade is No")
    else:
        print("The Evade is Yes")
```

```python
Bayes('Yes','Divorced',90)
Bayes('No','Married',60)
```
