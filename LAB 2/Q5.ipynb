```python
import pandas as pd
data = {'Name' : ['','Ali','Ahmed','Nida',''], 'Field' : ['C','E','E','C','C'],'Age': ['','','','',''],'Marks':[-90,60,-10,70,75]}
data = pd.DataFrame(data)
data
```

```python
data = data.drop(columns = 'Age')
data
```

```python
data = data.replace({'Name':{'' :'---'}})
data
```

```python
data = data.replace({'Field':{'C' :'0','E': '1'}})
data
```

```python
data['Marks'] = data['Marks'].where(data['Marks'] > 0)
data['Marks'].fillna(round(data['Marks'].mean(), 2), inplace = True)
data
```
